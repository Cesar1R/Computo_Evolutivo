```python
import ast  # Para convertir el contenido del archivo a un objeto Python (lista de tuplas)

def graficar_tuplas(ruta_archivo, path_pobIn, axs, k, color, title):
    """
    Lee un archivo con tuplas y las grafica en un gráfico específico de axs.
    
    :param ruta_archivo: Ruta del archivo que contiene las tuplas.
    :param axs: Lista de subgráficos (axes).
    :param k: Índice del subgráfico en axs donde se graficarán las tuplas.
    """
    # Abrir el archivo y leer su contenido
    with open(ruta_archivo, 'r') as archivo:
        # Leer todo el contenido del archivo y evaluarlo como una lista de tuplas
        contenido = archivo.read().strip()
        tuplas = ast.literal_eval(contenido)

    with open(path_pobIn, 'r') as ar:
        cont = ar.read().strip()
        tup = ast.literal_eval(cont) 

    axs[k].axvline(x = 2.2, color='black', linestyle='--')
    axs[k].axhline(y = 4.4, color='black', linestyle='--')
    axs[k].plot(2.2, 4.4, marker='o', color='black')

    for i, j in tuplas:
        axs[k].plot(i, j, marker='o', color=color)
    
    for i, j in tup: 
        axs[k].plot(i, j, marker='*', color='grey') 

    axs[k].set_title(title)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler

# Datos de ejemplo, organizados como en tu ciclo
hv_arr = [
    [4.22561, 7.55985, 14.438],      # Fila 1 (mutación 1.0/100, cruce 1.0/100)
    [16.5298, 8.9951, 4.57955],      # Fila 2 (mutación 10.0/100, cruce 1.0/100)
    [13.1248, 9.6535, 1.22093],      # Fila 3 (mutación 20.0/100, cruce 1.0/100)
    [3.00041, 7.61024, 12.9903],    # Fila 4 (mutación 30.0/100, cruce 1.0/100)
    [16.8682, 6.5986, 5.80692],     # Fila 5 (mutación 1.0/100, cruce 25.0/100)
    [6.87464, 5.22527, 2.09962],    # Fila 6 (mutación 10.0/100, cruce 25.0/100)
    [3.09529, 0.978589, 0.14527],   # Fila 7 (mutación 20.0/100, cruce 25.0/100)
    [13.9275, 7.75694, 4.57955],    # Fila 8 (mutación 30.0/100, cruce 25.0/100)
    [12.9903, 16.8682, 1.34904],    # Fila 9 (mutación 1.0/100, cruce 50.0/100)
    [1.55552, 3.26372, 2.77802],    # Fila 10 (mutación 10.0/100, cruce 50.0/100)
    [8.9951, 7.75694, 2.06542],     # Fila 11 (mutación 20.0/100, cruce 50.0/100)
    [4.57955, 7.55621, 2.22442],    # Fila 12 (mutación 30.0/100, cruce 50.0/100)
    [13.1248, 9.6535, 2.09148],     # Fila 13 (mutación 1.0/100, cruce 75.0/100)
    [1.22093, 5.73991, 0.455489],   # Fila 14 (mutación 10.0/100, cruce 75.0/100)
    [3.00041, 2.41755, -0.0295356], # Fila 15 (mutación 20.0/100, cruce 75.0/100)
    [7.61024, 2.58024, 1.85422],    # Fila 16 (mutación 30.0/100, cruce 75.0/100)
]

# Convertir los datos a un DataFrame de Pandas
mutaciones_arr = [1.0/100, 10.0/100, 20.0/100, 30.0/100]  # Columnas de mutación
cross_arr = [1.0/100, 25.0/100, 50.0/100, 75.0/100]  # Filas de cruce

# Reorganizar los datos en un array de 3 dimensiones (filas, columnas, funciones)
hv_arr_reorganizado = np.array(hv_arr).reshape(len(cross_arr), len(mutaciones_arr), 3)

# Convertimos esto en un DataFrame adecuado
df = pd.DataFrame(hv_arr_reorganizado.reshape(len(cross_arr)*len(mutaciones_arr), 3), 
                  columns=['WFG3', 'WFG2', 'WFG7'])

# Reshape de los índices para tener las combinaciones de cruce y mutación en un índice
df.index = pd.MultiIndex.from_product([cross_arr, mutaciones_arr], names=['Tasa de Cruce', 'Tasa de Mutación'])

# Normalizar los valores para cada función (WFG3, WFG2, WFG7) para que el máximo sea 1
scaler = MinMaxScaler()

# Normalizar cada función por separado
df_normalizado = df.copy()
df_normalizado['WFG3'] = scaler.fit_transform(df['WFG3'].values.reshape(-1, 1))
df_normalizado['WFG2'] = scaler.fit_transform(df['WFG2'].values.reshape(-1, 1))
df_normalizado['WFG7'] = scaler.fit_transform(df['WFG7'].values.reshape(-1, 1))

# Ahora generamos el mapa de calor para cada función

# Mapa de calor para WFG3
df_wfg3 = df_normalizado['WFG3'].unstack('Tasa de Mutación')
plt.figure(figsize=(10, 6))
sns.heatmap(df_wfg3, annot=True, cmap='YlGnBu', fmt='.2f', linewidths=.5)
plt.title("Mapa de Calor de Hipervolúmenes Normalizados para la Función WFG3", fontsize=16)
plt.ylabel("Tasa de Mutación", fontsize=12)
plt.xlabel("Tasa de Cruce", fontsize=12)
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.show()

# Mapa de calor para WFG2
df_wfg2 = df_normalizado['WFG2'].unstack('Tasa de Mutación')
plt.figure(figsize=(10, 6))
sns.heatmap(df_wfg2, annot=True, cmap='YlGnBu', fmt='.2f', linewidths=.5)
plt.title("Mapa de Calor de Hipervolúmenes Normalizados para la Función WFG2", fontsize=16)
plt.ylabel("Tasa de Mutación", fontsize=12)
plt.xlabel("Tasa de Cruce", fontsize=12)
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.show()

# Mapa de calor para WFG7
df_wfg7 = df_normalizado['WFG7'].unstack('Tasa de Mutación')
plt.figure(figsize=(10, 6))
sns.heatmap(df_wfg7, annot=True, cmap='YlGnBu', fmt='.2f', linewidths=.5)
plt.title("Mapa de Calor de Hipervolúmenes Normalizados para la Función WFG7", fontsize=16)
plt.ylabel("Tasa de Mutación", fontsize=12)
plt.xlabel("Tasa de Cruce", fontsize=12)
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.show()
```
